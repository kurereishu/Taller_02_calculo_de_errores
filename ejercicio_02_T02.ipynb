```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from matplotlib.animation import FuncAnimation, PillowWriter
from matplotlib.widgets import Slider, Button

# Variables simbólicas
x = sp.symbols('x')

# Ecuación: (x + 3)/2 = ±√2
soluciones_x = sp.solve((x + 3)/2 - sp.sqrt(2), x) + sp.solve((x + 3)/2 + sp.sqrt(2), x)
puntos = [(float(sol), -2) for sol in soluciones_x]

# Mostrar puntos de intersección
print("Puntos de intersección con y = -2:")
for p in puntos:
    print(f"x ≈ {p[0]:.4f}, y = {p[1]}")

# Gráfica
x_vals = np.linspace(-10, 5, 400)
X, Y = np.meshgrid(x_vals, x_vals)
lhs = (Y / 2)**2
rhs = ((X + 3) / 2)**2 - 1

# Configuración de la figura para la animación
fig, ax = plt.subplots(figsize=(12, 8))
plt.subplots_adjust(bottom=0.25)  # Dejar espacio para los widgets

contour = ax.contour(X, Y, lhs - rhs, levels=[0], colors='blue', label='Curva')
line, = ax.plot([], [], color='red', linestyle='--', label='y = -2')
scatter_points, = ax.plot([], [], 'go', zorder=5, label='Intersecciones')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Intersección de la curva con y = -2')
ax.legend()
ax.grid(True)
ax.set_aspect('equal')

# Widgets
ax_y = plt.axes([0.25, 0.1, 0.65, 0.03])
slider_y = Slider(ax_y, 'y', -10, 10, valinit=-2, valstep=0.1)

def update_plot(val):
    y = slider_y.val
    line.set_data(x_vals, [y] * len(x_vals))
    scatter_points.set_data([p[0] for p in puntos if p[1] == y], [y] * len([p for p in puntos if p[1] == y]))
    fig.canvas.draw_idle()

slider_y.on_changed(update_plot)

reset_button_ax = plt.axes([0.8, 0.025, 0.1, 0.04])
reset_button = Button(reset_button_ax, 'Reset')

def reset_plot(event):
    slider_y.reset()

reset_button.on_clicked(reset_plot)

# Crear la animación
ani = FuncAnimation(fig, update_plot, frames=np.linspace(-10, 10, 100), interval=50, blit=False)

# Guardar la animación como GIF
ani.save('interseccion_curva_y.gif', writer=PillowWriter(fps=20))

plt.show()
```
