```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Variables simbólicas
x = sp.symbols('x')

# Ecuación: (x + 3)/2 = ±√2
soluciones_x = sp.solve((x + 3)/2 - sp.sqrt(2), x) + sp.solve((x + 3)/2 + sp.sqrt(2), x)
puntos = [(float(sol), -2) for sol in soluciones_x]

# Mostrar puntos de intersección
print("Puntos de intersección con y = -2:")
for p in puntos:
    print(f"x ≈ {p[0]:.4f}, y = {p[1]}")

# Gráfica
x_vals = np.linspace(-10, 5, 400)
X, Y = np.meshgrid(x_vals, x_vals)
lhs = (Y / 2)**2
rhs = ((X + 3) / 2)**2 - 1

plt.contour(X, Y, lhs - rhs, levels=[0], colors='blue', label='Curva')
plt.axhline(-2, color='red', linestyle='--', label='y = -2')
plt.scatter(*zip(*puntos), color='green', zorder=5, label='Intersecciones')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Intersección de la curva con y = -2')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
